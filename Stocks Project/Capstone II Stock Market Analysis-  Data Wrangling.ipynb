# Capstone II Stock Market Analysis- Data Wrangling

This Capstone Project analyzes stock market data for some NASDAQ stocks.
We are going to analyze market stock data for Google, Amazon, Facebook, Apple and Microsoft.

Data is obtained from Yahoo's Finance API.


```python
#import analysis and visualization
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
from __future__ import division
```

```python
#to help grab stock data
#from pandas_datareader import the data
from pandas_datareader import data as pdr

import yfinance as yf
yf.pdr_override() # <== that's all it takes :-)

from datetime import datetime
```

```python
#reviewing performance over the last 12 months
#set end date to today
end = datetime.now()
```

```python
#set start date to 12 months back
start = datetime(end.year-1,end.month,end.day)
```

```python
#list of market stock data
market_list = ['GOOGL', 'AMZN', 'FB', 'AAPL', 'MSFT']
```

```python
#Yahoo Finance stock data
for stock in market_list:
    globals()[stock] = pdr.get_data_yahoo(stock, start, end)
    

```
